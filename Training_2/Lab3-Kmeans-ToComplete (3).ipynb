# K-Means for Iris Clustering

## Step 0 : Import necessary libraries

```python
# Import Numpy and Pandas
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

##  Step 1 :  Import & Split Iris Dataset

```python
#Import Data
iris = load_iris()

#Separate features and target variable
X = iris.data  
y = iris.target 
```

```python
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

#Print 10 first rows of Dataframe
iris_df.head(10)
```

```python
# Print Statistical Description of the dataframe
iris_df.describe()
```

## Step 2 : Initialize & Train K-Means Model

```python
# Initiate K mean Model
kmeans = KMeans(n_clusters=3)


#Run the model
kmeans.fit(X)
```

```python
# Print the coordinates of the clusters
print("Centres des clusters :")
print(kmeans.cluster_centers_)
```

## Step 4 : Make Predictions

```python
#Predict the Clusters 
y_pred = kmeans.predict(X)

# Print the Predicted labels
print(y_pred)
```

```python
iris_df.shape
```

```python
# Réduction des dimensions avec PCA pour visualiser en 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Visualizing K-means clusters in 2D space (PCA projection)
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_pred, cmap='viridis', marker='o', label='Clustered Data')

# Plotting the centroids of the clusters
centroids = kmeans.cluster_centers_
centroids_pca = pca.transform(centroids)  # Projecting centroids into 2D space
plt.scatter(centroids_pca[:, 0], centroids_pca[:, 1], c='red', s=200, marker='X', label='Centroids')

plt.title('K-means Clusters with Centroids')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```

```python
# Comparaison entre les labels réels et les labels prédits
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Confusion Matrix
cm = confusion_matrix(y, y_pred)

# Visualisation de la matrice de confusion
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.title('Matrice de confusion')
plt.xlabel('Prédictions')
plt.ylabel('Réel')
plt.show()
```

## Step 5 : Choose the K 

```python
# Elbow Method to determine the optimal number of clusters
inertia_values = []

for k in range(1, 11):  # Testing k values from 1 to 10
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia_values.append(kmeans.inertia_)

# Plotting the elbow graph
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia_values, marker='o', linestyle='-', color='b')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics.cluster import rand_score
from sklearn.metrics.cluster import adjusted_rand_score

ri = rand_score(y, y_pred)

# Adjusted Rand Index
ari = adjusted_rand_score(y, y_pred)

print("Rand Index: ", ri)
print("Adjusted Rand Index:",ari)
```
