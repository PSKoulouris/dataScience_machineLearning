# 2.3.2 Cross validation

# Question 1

```python
from sklearn.model_selection import KFold
```

# Question 2
# Logistic Regression

```python
kf10=KFold(n_splits=10, shuffle=False)
kf10.split(datanum)
```

# Questions 3 & 4

```python
i = 1
somme=0
for train_index, test_index in kf10.split(datanum):
    X_train = datanum.iloc[train_index].iloc[:, 0:10]
    X_test = datanum.iloc[test_index].iloc[:,0:10]
    y_train = datanum.iloc[train_index].loc[:,'y']
    y_test = datanum.loc[test_index]['y']
        
    #Train the model
    my_log.fit(X_train, y_train) #Training the model
    print(f"Accuracy for the fold no. {i} on the test set: {accuracy_score(y_test, my_log.predict(X_test))}")
    somme = somme + accuracy_score(y_test, my_log.predict(X_test))
    i += 1
accuracy=somme/10
accuracy
```

# Other solution

```python
from numpy import mean
from numpy import std
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
# prepare the cross-validation procedure
cv = KFold(n_splits=10, shuffle=False)
# evaluate model
scores = cross_val_score(my_log, dataX, dataY, scoring='accuracy', cv=cv, n_jobs=-1)
# report performance
print('Logistic regression accuracy: %.5f (%.5f)' % (mean(scores), std(scores)))
```
