# A simple SVM exercise 

```python
import sklearn# import library scikit learn
from sklearn.svm import SVC#import SVM classifiers
import pandas as pd
```

```python
dataset = [[6,9.5,-1],[0.8,9.2,-1],[1.8,6.5,-1],[4.5,6.5,-1],[7,5.5,1],[4.5,4,1],[9,3,1],[2.5,1.5,1],[5.5,0.5,1] ]
```

```python
df=pd.DataFrame(dataset,index=["E1", "E2","E3","E4","E5","E6","E7","E8", "E9"], 
columns=["X_1","X_2","Label"])
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter('X_1','X_2', data=df, c='Label')
```

```python
from sklearn.svm import SVC
```

```python
X = df.drop('Label',axis=1)

y = df['Label']
```

```python
clf_train = SVC(kernel="linear")
clf_train.fit(X, y) 
```

```python
pred = clf_train.predict(X)
import sklearn.metrics
from sklearn.metrics import accuracy_score
score = sklearn.metrics.accuracy_score(y, pred)
print(score)
```
